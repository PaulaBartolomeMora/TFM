```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Código de sampleo para reducir volumen de datos: creación de nueva columna H, obtención de media de valores/hora y agrupación de filas por horas

```python
dataset = 'dataset/preprocessed/new_power_samples_d3_1split_1.csv'
df = pd.read_csv(dataset)  
df.head()
```

## PARÁMETROS INPUT

```python
DATETIME_SAMPLE = '2013-01-01'
HOUR_SAMPLE = 0
IID_SAMPLE = 24
```

```python
df['tmstp'] = pd.to_datetime(df['tmstp'])
df['H'] = df['tmstp'].dt.strftime('%H') #nueva columna de hora
df['H'] = df['H'].astype(int)
```

```python
iid_array = df['iid'].unique() #ids nodos que aparecen en el dataset
H_array = df['H'].unique() #array de horas

print(iid_array) 
print(H_array) 
```

```python
#for id in iid_uniques:
#    for h in H_uniques:

filter = df[(df['iid'] == IID_SAMPLE) & (df['H'] == HOUR_SAMPLE) & (df['datetime'] == DATETIME_SAMPLE)] #filtro parámetros input
filter.head()
```

```python
X = filter.iloc[:, 3:18] #se extraen columnas features -> nombre + valores
X_features = X.columns.to_list()
X_values = X.values

print(X_features) #nombres features 
print(X_values) #valores features 
```

```python
hourly_data = pd.DataFrame(columns=X_features) #df de almacenamiento
array = np.array([])

for feature in X_features:
    data = filter[feature]
    print(data.mean())
    temp = pd.DataFrame({feature: [data.mean()]})
    hourly_data = pd.concat([hourly_data, temp], ignore_index=True)

hourly_data.to_csv('OUT.csv', index=False)
```

CÃ³digo de merge datasets environment y production y parseo timestamp (nuevas columnas -> datetime, hour (h,m,s))

```python
df_env = pd.read_csv('dataset/environment.csv')
df_prod = pd.read_csv('dataset/production.csv')
```

### Timestamp processing

```python
df_env['TIMESTAMP'] = pd.to_datetime(df_env['TIMESTAMP'])
df_env.head(10)
```

```python
df_prod = df_prod.rename(columns={'datetime': 'TIMESTAMP'})
df_prod['TIMESTAMP'] = pd.to_datetime(df_prod['TIMESTAMP'])
df_prod.head()
```

```python
df_merged = pd.merge(df_env, df_prod, on='TIMESTAMP')
df_merged['date'] = df_merged['TIMESTAMP'].dt.strftime('%Y-%m-%d')
df_merged['hour'] = df_merged['TIMESTAMP'].dt.strftime('%H:%M:%S')

df_merged.head()
df_merged.to_csv('dataset\env+prod.csv', index=False)
```

## DATASET FINAL DE CARGAS PARA SIMULACIÓN

```python
import pandas as pd
import glob as gl
```

```python
#parámetros timestamp y fichero de salida
fecha = '2010-11-28' 
hora = 0 
out = "simtests/test_" + str(fecha) + "_" + str(hora)
```

```python
#lectura de todos los dfs
path = gl.glob("dataset/nodes/loads/load_*.csv")
dfs = []

for file in path:
    dfs.append(pd.read_csv(file, low_memory=False))
    
df = pd.concat(dfs, ignore_index=True)
```

### Selección de instante temporal [2010-11-28, 2011-11-28)

```python
subset = df[(df['datetime'] == fecha) & (df['h'] == hora)] 
subset.count() #tienen que ser 23 datos
```

```python
column_order = ['iid'] + [col for col in list(subset.columns) if col not in ['iid']] #reorden de columna iid
subset = subset[column_order]
subset = subset.sort_values(by = 'iid')
subset
```

```python
if subset.shape[0] == 23 and not subset.empty: #23 filas con datos
    subset.to_csv(out, index=False)
```

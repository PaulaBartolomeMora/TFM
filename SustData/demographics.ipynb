Conversión de demographics.xlsx a .csv y gráfico de rangos temporales de los despliegues de nodos

```python
df = pd.read_excel('demographics.xlsx')
df.to_csv('demographics.csv', index=False)
df = pd.read_csv('demographics.csv')  
```

```python
df['Start'] = pd.to_datetime(df['Start'])
df['End'] = pd.to_datetime(df['End'])
df.head(5)
```

```python
plt.figure(figsize=(20, 6))
df.set_index('IID', inplace=True)

for i, row in df.iterrows():
    plt.plot([row['Start'], row['End']], [i, i])
    plt.text(row['Start'], i, f'{i}', ha='right', fontsize=8)

plt.xlabel('Timestamp')
plt.ylabel('iid')
plt.title('DEPLOYMENT TEMPORAL RANGES')
plt.legend()
plt.show()
```

Conversión de demographics.xlsx a .csv y gráfico de rangos temporales de los despliegues de nodos

```python
#df = pd.read_excel('dataset/demographics.xlsx')
#df.to_csv('dataset/demographics.csv', index=False)
df = pd.read_csv('dataset/demographics.csv')  
```

```python
df['Start'] = pd.to_datetime(df['Start'])
df['End'] = pd.to_datetime(df['End'])
df.head(23)
```

```python
plt.figure(figsize=(15, 7))
df.set_index('IID', inplace=True)

for i, row in df.iterrows():
    plt.plot([row['Start'], row['End']], [i, i])
    plt.text(row['Start'], i, f'{i}', ha='right', fontsize=8)

plt.xlabel('Timestamp')
plt.ylabel('iid')
plt.title('Períodos temporales de los despliegues de SustDataED')
plt.xlim(pd.Timestamp('2010-06-20'),pd.Timestamp('2014-04-01'))
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
df.set_index('IID', inplace=True)

# Limitar a las primeras 23 filas
df_subset = df.head(23)

for i, row in df_subset.iterrows():
    plt.plot([row['Start'], row['End']], [i, i])
    plt.text(row['End'], i, f'{i}', ha='left', va='center', fontsize=8)

plt.xlabel('Timestamp')
plt.ylabel('IID')
plt.title('Período temporal del segundo despliegue de SustDataED')
plt.grid(True)
plt.show()
```
